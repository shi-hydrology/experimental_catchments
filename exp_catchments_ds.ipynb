```python
import pandas as pd
```

```python
data = pd.read_csv("./src/data/1_daily.csv")
data['DATE_TIME'] =  pd.to_datetime(data['DATE_TIME'], format='%Y-%m-%d')
data
```

```python
dict_data={}
for header in data.columns.values.tolist():
    dict_data[header]=data[header].tolist()
dict_data
```

```python
import json
jsn=json.dumps(dict_data)
jsn
```

```python
with open('./daily.json', 'w') as file:
    file.write(jsn)
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10,6)

names={'MEAN_DAILY_AIR_TEMP_CELSIUS':'  , ', 
       'MAX_DAILY_AIR_TEMP_CELSIUS':'  , ', 
       'MIN_DAILY_AIR_TEMP_CELSIUS':'  , ', 
       'SOIL_MIN_TEMP_CELSIUS':'  , ',
       'PRECIPITATION_MM':' , ', 
       'SNOW_THICKNESS_CM':'  , '}

for key in names.keys():
    plt.plot(data['DATE_TIME'], data[key], label=names[key])
    plt.grid()
    plt.legend()
    #plt.savefig('./src/data/'+key+'.png', dpi=350)
    plt.show()
```
